### 테이블 수정 (날짜 집어넣기)

```python
import pandas as pd
```

```python
df = pd.read_excel('ppl_all_product.xlsx', encoding='utf8-sig')
```

```python
product = []
a = -30
for i in range(0,len(df[df.columns[0]])):
    b = a+i
    if b < 0:
        product.append('D'+str(a+i))
    elif b == 0:     
        product.append('D-day')
    else:
        product.append('D+'+str(a+i))
product
```

```python
conn = pymysql.connect(host = 'teamproject.cq3n19tmwx8i.us-east-2.rds.amazonaws.com',user='bigdb',password='bigdb1234', \
                      db='ppl',charset='utf8')
```

```python
cur = conn.cursor(pymysql.cursors.DictCursor)
```

```python
for i in product:
    sql = "ALTER TABLE `product_all` ADD COLUMN `"+i+"` FLOAT NOT NULL DEFAULT 0 "
    cur.execute(sql)
    conn.commit()
```

```python
cur.close()
conn.close()
```
